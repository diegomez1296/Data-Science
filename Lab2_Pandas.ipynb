```python
##Merge
#Concat
#Łączenie obiektów pandas
df = pd.DataFrame(np.random.randn(10, 4))
print(df)
print("____________")
pieces = [df[:3], df[3:7], df[7:]]
print(pieces)
print("____________")
pd.concat(pieces)
```

```python
#Join
#Łączenie danych metodą zaczerpniętą z SQL 
left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})

print(left)
print("____________")
print(right)
print("____________")
print(pd.merge(left, right, on='key'))
```

```python
#Append
#Dodanie na końcu
df = pd.DataFrame(np.random.randn(8, 4), columns=['A', 'B', 'C', 'D'])
print(df)
print("____________")
s = df.iloc[3]
print(s)
print("____________")
df.append(s, ignore_index=True)
```

```python
##Grouping
#groupby -> -Podział danych na grupy na podstawie niektórych kryteriów
#           -Zastosowanie funkcji do każdej grupy niezależnie
#           -Łączenie wyników w strukturę danych

df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                   'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                    'C': np.random.randn(8),
                      'D': np.random.randn(8)})
df
```

```python
df.groupby('A').sum()
```

```python
df.groupby(['A', 'B']).sum()
```

```python
##Reshaping
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz','foo', 'foo', 'qux', 'qux'],
                   ['one', 'two', 'one', 'two','one', 'two', 'one', 'two']]))

index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
df2 = df[:4]
df2
```

```python
#Stack
#Metoda stack () „kompresuje” poziom w kolumnach DataFrame.
stacked = df2.stack()
stacked
```

```python
#Operacja odwrotna do stack()
stacked.unstack(1)
```

```python
#Pivot tables
df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 3,
                       'B': ['A', 'B', 'C'] * 4,
                       'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                       'D': np.random.randn(12),
                       'E': np.random.randn(12)})
    

df
```

```python
#Za pomocą pivot_table możemy bardzo łatwo tworzyć tabele przestawne:
pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])
```

```python
##Time series
rng = pd.date_range('1/1/2012', periods=100, freq='S')
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts.resample('5Min').sum()
```

```python
#Time zone representation:
rng = pd.date_range('3/6/2012 00:00', periods=5, freq='D')
ts = pd.Series(np.random.randn(len(rng)), rng)
ts
```

```python
ts_utc = ts.tz_localize('UTC')
ts_utc
```

```python
#Konwersja na inną stefę czasową:
ts_utc.tz_convert('US/Eastern')
```

```python
#Konwersja między reprezentacjami przedziałów czasowych:
rng = pd.date_range('1/1/2012', periods=5, freq='M')
ts = pd.Series(np.random.randn(len(rng)), index=rng)

print(ts)
print("____________")
ps = ts.to_period()
print(ps)
print("____________")
ps.to_timestamp()
```

```python
prng = pd.period_range('1990Q1', '2000Q4', freq='Q-NOV')

ts = pd.Series(np.random.randn(len(prng)), prng)

ts.index = (prng.asfreq('M', 'e') + 1).asfreq('H', 's') + 9

ts.head()
```

```python
##Categoricals
#Konwersja surowych ocena na kategoryczny typ danych
df = pd.DataFrame({"id": [1, 2, 3, 4, 5, 6],"raw_grade": ['a', 'b', 'b', 'a', 'a', 'e']})
print(df)
print("____________")
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

```python
#Zamiana nazw kategorii na bardziej znaczące
df["grade"].cat.categories = ["very good", "good", "very bad"]
df["grade"]
```

```python
df.sort_values(by="grade")
```

```python
import matplotlib.pyplot as plt
##Plotting
#Generowanie wykesu
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```
