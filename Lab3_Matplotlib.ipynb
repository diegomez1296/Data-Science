```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
cars = pd.read_csv('samochody1tys.csv')
```

```python
##Wykresy
x = np.linspace(-1, 1, 100)
y = np.tan(x)
plt.plot(x, x, label='liniowa - f(x)')
plt.plot(x, x**2, label='kwadratowa - f(x^2)')
plt.plot(x, x**3, label='trzeciego rzędu - f(x^3)')
plt.plot(x, y, label='trygonometryczna - tg(x)')

plt.xlabel('Oś X')
plt.ylabel('Oś Y')

plt.title("Mój wykres")

plt.legend()

plt.show()
```

```python
##Dane do wykresu słupkowego
#Suma samochodów danej marki, których ilość przekroczyła 5% pliku samochody
c = cars.iloc[:, 0:2]
c_group = c.groupby('marka').count()
c_copy = c_group.copy()
c_5percent = c_copy[c_copy["id"] >= 50]
print(c_5percent) #marki posiadające 5% tabeli cars

numbersOfBars = c_5percent.iloc[:, 0:2].count()["id"]
numbersOfBars
```

```python
##Wykres słupkowy
#Przedstawienie ilości marek samochodów, których ilość przekroczyła 5% wszystkich samochodów w pliku
N = np.arange(numbersOfBars)
carsAmount = c_5percent["id"]
width = 0.8
p = plt.bar(N, carsAmount, width, 0)
plt.ylabel('Amount')
plt.title('Cars 5% plot')
plt.xticks(N, ('Audi','BMW','Ford','Mercedes','Opel','Renault','VW'))

plt.show()
```

```python
##Dane do wykresu kołowego
#Suma ilości każdego rodzaju silnika w samochodach
c2 = cars.iloc[:, [0,4]]
c2_group = c2.groupby('rodzaj_silnika').count()
print(c2_group)
```

```python
##Wykres kołowy
#Przedstawia procentowy rozkład silników w autach
labels = 'Benzyna', 'Benzyna+CNG', 'Benzyna+LPG', 'Diesel ','Elektryczny', 'Hybryda'
sizes = c2_group["id"]
explode = (0.1, 0.7, 0.7, 0.1, 0.8, 0.5)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=0)
ax1.axis('equal')

plt.show()
```

```python
##Dane do histogramu
#Ilość aut wyprodukowanych w danym roku
c3 = cars.iloc[:, [0,3]]
c3_group = c3.groupby('rok_produkcji').count()
print(c3_group)
```

```python
##Histogram
#Wyciągnięcie z pliku kolumny 'rok_ produkcji' i zamienienie jej na tablicę
carYear = cars.iloc[:, [3]].to_numpy()
bins = 20
plt.style.use('fivethirtyeight')
plt.hist(carYear, bins= bins, edgecolor = 'black', color='#0504aa', alpha=0.7, rwidth=0.9)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Rok produkcji')
plt.ylabel('Ilość')
plt.title('Auta')
```
