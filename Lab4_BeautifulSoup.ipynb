```python
from openpyxl import Workbook
wb = Workbook()
ws = wb.active
```

```python
#Tworzenie nowych arkuszy
ws1 = wb.create_sheet("Giełda")
ws2 = wb.create_sheet("Linki") 
ws3 = wb.create_sheet("Filmweb") 
print(wb.sheetnames)
```

```python
from random import choice
from string import ascii_lowercase
import requests
from bs4 import BeautifulSoup
#Generowanie kodów
def generateCode(n):
    return "".join(choice(ascii_lowercase) for i in range(n))

codes = [generateCode(3) for _ in range(50)]

valid_codes = []
#Sprawdzanie poprawności wygenerowanych kodów
for code in codes:
    if len(valid_codes) < 5:
        print(code)
        params = {"s": code}
        response = requests.get("http://stooq.pl/q/", params = params)
        soup = BeautifulSoup(response.content, 'html.parser') 
        #print(soup.prettify())
        if len(soup.body.findAll('span', id='c2')) != 0:
            try:
                link_content = soup.find('font',id='f16').find('a',href=True)
            except:
                  print("An exception occurred") 
            if link_content.text == 'symbol waloru':
                print("Brak kodu")
            else:
                valid_code = link_content['href'][5:]
                valid_codes.append(valid_code)
                print(f'Znalezionio podobny kod: {valid_code}')
        else:
            valid_codes.append(code)
```

```python
ws1.cell(row=1, column=1).value = 'Kod'
ws1.cell(row=1, column=2).value = 'Kurs'
ws1.cell(row=1, column=3).value = 'Zmiana procentowa'
ws1.cell(row=1, column=4).value = 'Transakcje'
row = 2
#Zapis do arkusza wybranych danych z notowań giełdowych, których kod został znaleziony
for code in valid_codes:
    print(code)
    params = {"s": code}
    response = requests.get("http://stooq.pl/q/", params=params)
    soup = BeautifulSoup(response.content, 'html.parser')
    price = soup.find(text="Kurs").parent.find("span").text
    percent_change = soup.find(text="Zmiana").parent.findAll("span")[1].text
    transactions = soup.find(text="Transakcje").parent.find("span").text
    print(f"{price}, {percent_change}, {transactions}")
    ws1.cell(row=row, column=1).value = code
    ws1.cell(row=row, column=2).value = price if price else 'null'
    ws1.cell(row=row, column=3).value =  percent_change if percent_change else 'null'
    ws1.cell(row=row, column=4).value =  transactions if transactions else 'null'
    row += 1
```

```python
from bs4 import BeautifulSoup
import requests
#Wyciąganie linków ze strony
headers = requests.utils.default_headers()
url = "https://assetstore.unity.com/";
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
#Wyszukiwanie linków
links = [a['href'] for a in soup.find_all('a', href=True) if a.text]

#Wyświetlenie i zapis do arkusza linków znajdujących się w tabeli links
row= 1
for link in links:
    ws2.cell(row, column = 1).value = link
    print(link);
    row +=1
```

```python
url = "https://www.filmweb.pl/Desperado"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')

filmweb = {}
filmweb['Director'] = soup.find(text="reżyseria:").next_element.find("a").text
filmweb['Year'] = soup.find("h1", class_="filmTitle").find('span').text
filmweb['BoxOffice']  = soup.find(text="boxoffice:").next_element.text
filmweb['Raiting'] = soup.find("span", itemprop="ratingValue").text

row = 1
for key, value in filmweb.items():
    ws3.cell(row,column=1).value = key
    ws3.cell(row,column=2).value = value
    print(key)
    print(value)
    row += 1

```
